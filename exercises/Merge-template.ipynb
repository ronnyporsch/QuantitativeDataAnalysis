```python
# What happens if there are 2 identical columns?
data1 = {
    'id': [1, 2, 3, 4, 5],
    'A': np.random.randint(1, 100, size=5),
    'B': np.random.randint(1, 100, size=5),
    'C': np.random.randint(1, 100, size=5),
    'D': np.random.randint(1, 100, size=5),
    'E': np.random.randint(1, 100, size=5)
}
df1 = pd.DataFrame(data1)

# Create second data record
data2 = {
    'id': [1, 2, 3, 4, 5],
    'A': np.random.randint(1, 100, size=5), # A occurs in both data sets
    'G': np.random.randint(1, 100, size=5),
    'H': np.random.randint(1, 100, size=5),
    'I': np.random.randint(1, 100, size=5),
    'J': np.random.randint(1, 100, size=5)
}
df2 = pd.DataFrame(data2)

# Link data using the "id" column
merged_df = pd.merge(df1, df2, on='id', how='inner')

# Show results
print("data1")
print(df1)
print("\ndata2:")
print(df2)
print("\nmerged data:")
print(merged_df)
```

```python
# What if not all ids match?
data1 = {
    'id': [1, 2, 3, 4, 6],
    'A': np.random.randint(1, 100, size=5),
    'B': np.random.randint(1, 100, size=5),
    'C': np.random.randint(1, 100, size=5),
    'D': np.random.randint(1, 100, size=5),
    'E': np.random.randint(1, 100, size=5)
}
df1 = pd.DataFrame(data1)

# Create second data record
data2 = {
    'id': [1, 2, 3, 4, 5],
    'F': np.random.randint(1, 100, size=5), # A occurs in both data sets
    'G': np.random.randint(1, 100, size=5),
    'H': np.random.randint(1, 100, size=5),
    'I': np.random.randint(1, 100, size=5),
    'J': np.random.randint(1, 100, size=5)
}
df2 = pd.DataFrame(data2)

# Link data using the "id" column
merged_df = pd.merge(df1, df2, on='id', how='inner')

# Show results
print("data1")
print(df1)
print("\ndata2:")
print(df2)
print("\nmerged data:")
print(merged_df)
```

```python
# We can change "how=" to "outer", then the ids will still be transferred
# This means we do not lose any data
merged_df = pd.merge(df1, df2, on='id', how='outer')

# Show results
print("data1")
print(df1)
print("\ndata2:")
print(df2)
print("\nmerged data:")
print(merged_df)
```
